```python
print("Kütüphane Otomasyon Uygulaması Başlıyor")
print("******************************************")
def menu():
  print("******************************************")
  print("[1] Kitapları listele")
  print("[2] Kitap ekle")
  print("[0] Uygulamayı kapat")
  print("******************************************")
  option = int(input("Yapacağınız işlemi seçin:"))
  if option == 1 :
    listele()
    menu()
  elif option== 2:
    print("Kitap eklemek için gerekli bilgiler alınıyor...")
    eserAdi = input("Eser adını giriniz:")
    yazarAdi= input("Yazar adını giriniz:")
    yayinEvi= input("Yayınevini giriniz:")
    bTarihi= input("Basım tarihini giriniz:")
    isbn=input("ISBN kodunu giriniz:")
    ekle(eserAdi, yazarAdi, yayinEvi, bTarihi, isbn)
    menu()
  elif option == 0:
    print("Uygulamadan çıkılıyor")
  else:
    print("Lütfen geçerli bir seçenek giriniz.")


def ekle(eser, yazar, yayinevi, basimTarihi, isbn ):
  print("Kitap ekleniyor.")
  yeniEser = "%s %s %s %s %s \n" % (eser, yazar, yayinevi, basimTarihi, isbn)
  print(yeniEser)
  dosya = open("final.txt", "a")
  dosya.write(yeniEser)
  dosya.close()
  print("Kitap eklendi.")

def listele():
  print("Kitaplar listeleniyor...")
  dosya = open("final.txt", "r")
  print(dosya.read())
  dosya.close()


menu()
```
